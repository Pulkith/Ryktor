```python
import os
import pandas as pd
```

```python
directory_name = "/Users/pulkith/Downloads/Physician Office Visit Costs"
# find all files in the directory
files = os.listdir(directory_name)
# load each file into a df
dfs = {}
for file in files:
    if file.endswith(".csv"):
        dfs[file[:-4]] = pd.read_csv(os.path.join(directory_name, file))
    elif file.endswith(".xlsx"):
        print("FOUND")
        dfs[file] = pd.read_excel(os.path.join(directory_name, file))
    else:
        print(f"Skipping file {file} because it is not a CSV or Excel file")

# for each df, add a column, at the very start, that is the filename for each row
for key, df in dfs.items():
    df.insert(0, "filename", key)
```

```python
# concat all the dfs into one

df = pd.concat(dfs.values(), ignore_index=True)
```

```python
# save the df to a csv
df.to_csv("insurance_copay_splits.csv", index=False)
```
